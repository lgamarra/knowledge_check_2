Clean up #1, update category name from "City" to "CITY"

clean up #2, the zip code column has a ".0" after every zip code. Created a code to get rid of that. 

```python
def clean_zipcode(zipcode):
    if pd.notnull(zipcode):
        zipcode = str(zipcode).split(".")[0]
        return zipcode.strip()
    else:
        return zipcode

lmpd_df['ZIP_CODE'] = lmpd_df['ZIP_CODE'].apply(clean_zipcode)
```
